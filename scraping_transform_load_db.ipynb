```python
# Join id wilayah between LPSE and Tender
df_pengumuman = pd.merge(tender_src[['id_duplikat', 'satker', 'nilai_pagu', 'lokasi_pekerjaan', 'cara_bayar', 'jumlah_peserta', 'domain']],
                        lpse_src[['id_wilayah', 'domain']], how='left', left_on='domain',right_on='domain')
df_pengumuman.drop('domain', axis=1, inplace=True)

# Province column
df_pengumuman['id_wilayah'] = df_pengumuman.id_wilayah.astype('string')
df_pengumuman['id_wilayah_provinsi'] = df_pengumuman['id_wilayah'].str[:2]
df_pengumuman['id_wilayah'] = df_pengumuman.id_wilayah.astype(int)
df_pengumuman['id_wilayah_provinsi'] = df_pengumuman.id_wilayah_provinsi.astype(int)

# Cleaning Package name
list_kolom = ['satker', 'lokasi_pekerjaan']
for i in list_kolom:
    tender_src[i] = tender_src[i].str.strip(', -')

# Join id wilayah between Pengumuman and Wilayah Table (provinces and disticts columns)
tabel_pengumuman = pd.merge(df_pengumuman, tabel_wilayah, how='left', left_on='id_wilayah', right_on='id_wilayah')
tabel_pengumuman = pd.merge(tabel_pengumuman, tabel_wilayah, how='left', left_on='id_wilayah_provinsi', right_on='id_wilayah')

# Rename and Remove Unnecessary Columns
tabel_pengumuman.drop(['id_wilayah_x', 'id_wilayah_provinsi', 'id_wilayah_y'], axis=1, inplace=True)
tabel_pengumuman.rename(columns={'wilayah_x':'kab_pekerjaan', 'wilayah_y':'prov_pekerjaan'}, inplace=True)
nama_kolom = ['id_duplikat', 'satker', 'nilai_pagu', 'lokasi_pekerjaan', 
              'kab_pekerjaan', 'prov_pekerjaan', 'cara_bayar', 'jumlah_peserta']
tabel_pengumuman = tabel_pengumuman.reindex(columns=nama_kolom)
tabel_pengumuman.rename({'id_duplikat':'id_tender'}, axis=1, inplace=True)

tabel_pengumuman[tabel_pengumuman.duplicated(subset='id_tender', keep=False)].sort_values(by=['id_tender']).reset_index(drop=True)
```

```python
def type_detection(detail_kualifikasi):
    if ("Kualifikasi Teknis" in str(detail_kualifikasi) and "Kualifikasi Administrasi" in str(detail_kualifikasi)) or ("Teknis" in str(detail_kualifikasi) and "Administrasi" in str(detail_kualifikasi)):
        return "Semua"
    elif "Kualifikasi Teknis" in str(detail_kualifikasi) or "Teknis" in str(detail_kualifikasi):
        return "Teknis"
    elif "Kualifikasi Administrasi" in str(detail_kualifikasi) or "Administrasi" in str(detail_kualifikasi):
        return "Administrasi"
    else:
        return "Semua"
syarat_kualifikasi["kategori"] = syarat_kualifikasi.detail_kualifikasi.apply(lambda x: type_detection(x))
```
