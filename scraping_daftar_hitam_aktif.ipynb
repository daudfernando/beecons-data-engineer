```python
npwp_list = []
nama_list = []
sk_penetapan_list = []
pelanggaran_list = []
klpd_list = []
satker_list = []
masa_berlaku_sanksi_list = []
tanggal_tayang_list = []
status_list = []
```

```python
# set perulangan halaman daftar hitam
for q in tqdm(range(49,50)):
    # set perulangan untuk list peserta daftar hitam (per halaman, lazimnya, ada 10)
    for i in tqdm(range(2, 10)):
        url_daftar_hitam = 'https://inaproc.id/daftar-hitam?page='+str(q)
        driver.set_window_size(1920,1080)
        driver.get(url_daftar_hitam)

        # cek tombol untuk buka pop up page
        button = driver.find_element(By.XPATH, '//*[@id="blacklist"]/div[2]/div/section/div[3]/div[1]/table/tbody/tr[' + str(i) + ']/td[1]/h5/a')
        driver.execute_script("arguments[0].click();", button)
        time.sleep(1)

        # menghitung ada berapa pelanggaran peserta (dalam satu pop up page)
        total_violation = len(driver.find_elements(By.XPATH, '//*[@id="injunctions"]/tbody/tr'))
        print(total_violation)

        # perulangan untuk setiap list pelanggaran peserta yang ada
        for i in range(1, total_violation+1):
            # cek untuk setiap elemen menggunakan xpath
            npwp = driver.find_element(By.XPATH, '//*[@id="npwp"]')
            npwp_list.append(npwp.text.strip())
            
            nama = driver.find_element(By.XPATH, '//*[@id="nama-penyedia"]')
            nama_list.append(nama.text.strip())

            sk_penetapan = driver.find_element(By.XPATH, '//*[@id="injunctions"]/tbody/tr[' + str(i) + ']/td[1]')
            sk_penetapan_list.append(sk_penetapan.text.replace('No : ', '').strip())

            pelanggaran = driver.find_element(By.XPATH, '//*[@id="injunctions"]/tbody/tr[' + str(i) + ']/td[2]/div/div/div')
            pelanggaran_list.append(pelanggaran.text.strip())

            klpd = driver.find_element(By.XPATH, '//*[@id="injunctions"]/tbody/tr[' + str(i) + ']/td[3]/table/tbody/tr[1]/td[2]')
            klpd_list.append(klpd.text.strip())

            satker = driver.find_element(By.XPATH, '//*[@id="injunctions"]/tbody/tr[' + str(i) + ']/td[3]/table/tbody/tr[2]/td[2]')
            satker_list.append(satker.text.strip())

            masa_berlaku_sanksi = driver.find_element(By.XPATH, '//*[@id="injunctions"]/tbody/tr[' + str(i) + ']/td[3]/table/tbody/tr[3]/td[2]')
            masa_berlaku_sanksi_list.append(masa_berlaku_sanksi.text.strip())

            tanggal_tayang = driver.find_element(By.XPATH, '//*[@id="injunctions"]/tbody/tr[' + str(i) + ']/td[3]/table/tbody/tr[4]/td[2]')
            tanggal_tayang_list.append(tanggal_tayang.text.strip())
            
            status = driver.find_element(By.XPATH, '//*[@id="injunctions"]/tbody/tr[' + str(i) + ']')
            cek = str(status.get_attribute('outerHTML'))
            if '<tr class="active">' in cek:
                # status satu artinya aktif
                status_list.append('1')
            else:
                # status 0 artinya habis masa berlaku
                status_list.append('0')
```
